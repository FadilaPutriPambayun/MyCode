````python
hari_df["dteday"] = pd.to_datetime(hari_df["dteday"])
hari_df.info()
````

````python
hari = hari_df.groupby(by="season").agg({
    "instant": "nunique",
    "cnt": ["max", "min", "mean", "std"]
})
print("Demograsi Tabel Hari :", hari)
````

````python
jam = jam_df.groupby(by="season").agg({
     "instant": "nunique",
    "cnt": ["max", "min", "mean", "std"]
})
print("Demograsi Tabel jam :", jam)
````



*   Berdasarkan tabel diatas demografi penyewa sepeda harian Musim ke-3 adalah musim dengan penyewaan harian tertinggi, dengan minimal 1.115 dan max 8.714
*   Item daftar



````python
# Melihat Demografi Penyewa Sepeda Berdasarkan Cuaca
hari = hari_df.groupby(by="weathersit").agg({
    "instant": "nunique",
    "cnt": ["max", "min", "mean", "std"]
})
print("Demografi Tabel Hari: ", hari)
````

````python
jam = jam_df.groupby(by="weathersit").agg({
    "instant": "nunique",
    "cnt": ["max", "min", "mean", "std"]
})
print("Demografi Tabel jam: ", jam)
````

````python
# Persebaran Data Hari
print('Data hari_df')
print('tabel data hari :', hari_df.groupby(by="holiday").cnt.nunique().sort_values(ascending=False))
print('tabel data hari :', hari_df.groupby(by="weekday").cnt.nunique().sort_values(ascending=False))
print('tabel data hari :', hari_df.groupby(by="workingday").cnt.nunique().sort_values(ascending=False))
print('tabel data hari :', hari_df.groupby(by="yr").cnt.nunique().sort_values(ascending=False))
````

````python
# Persebaran Data Jam
print('Data jam_df')
print('tabel data jam :', hari_df.groupby(by="holiday").cnt.nunique().sort_values(ascending=False))
print('tabel data jam :', hari_df.groupby(by="weekday").cnt.nunique().sort_values(ascending=False))
print('tabel data jam :', hari_df.groupby(by="workingday").cnt.nunique().sort_values(ascending=False))
print('tabel data jam :', hari_df.groupby(by="yr").cnt.nunique().sort_values(ascending=False))
````

  Visualisasi Data

````python
# Visualisasi Data Terhadap Season
plt.figure(figsize=(36,12))
plt.subplot(1,2,1)
ax = sns.barplot(x='season', y='cnt', data=hari_df, palette='Paired')
plt.title("Korelasi antara musim dengan jumlah penyewaan")
ax.bar_label(ax.containers[0], fmt= '%0.if', label_type='center')
plt.subplot(1,2,2)
plt.title("Korelasi antara musim dengan jumlah penyewaan, dengan hue = yr")
ax = sns.barplot(x='season', y='cnt', hue='yr', data=hari_df, palette='Paired')
````

````python
# Visualisasi Data Terhadap Bulan
plt.figure(figsize=(36,12))
plt.subplot(2,3,1)
plt.title("Korelasi antara bulan dengan cnt")
ax = sns.barplot(x='mnth', y='cnt', data=hari_df, palette='Paired')
ax.bar_label(ax.containers[0], fmt= '%0.if', label_type='center')
plt.subplot(2,3,2)
plt.title("Korelasi antara bulan dengan cnt, dengan hue = yr")
ax = sns.barplot(x='mnth', y='cnt', hue= 'yr', data=hari_df, palette='Paired')
````

````python
# Visualisasi Data Terhadap Cuaca
plt.figure(figsize=(36,12))
plt.subplot(2,3,1)
plt.title("Korelasi antara weathersit dengan cnt")
ax = sns.barplot(x='weathersit', y='cnt', data=hari_df, palette='Paired')
ax.bar_label(ax.containers[0], fmt= '%0.if', label_type='center')
plt.subplot(2,3,2)
plt.title("Korelasi antara weathersit dengan cnt, dengan hue = yr")
ax = sns.barplot(x='weathersit', y='cnt', hue= 'yr', data=hari_df, palette='Paired')
````
